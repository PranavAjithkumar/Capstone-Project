```python
for lat, lng, City in zip(dfsrt['LATITUDE'], dfsrt['LONGITUDE'], dfsrt['CityName']):    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&categoryId={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        catId,
        lat,
        lng,       
        radius, 
        LIMIT      
        )
    
    # make the GET request
    result = requests.get(url)     
    if result.status_code == 200:
        results = json.loads(result.content.decode())
        Rvenues = results['response']['groups'][0]['items']
        
    for venue in Rvenues:
        venues.append((
            City,
            lat, 
            lng, 
            venue['venue']['name'],          
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],           
            venue['venue']['categories'][0]['name']))
  
```

```python
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)
# define the column names
venues_df.columns = ['CityName', 'Latitude', 'Longitude', 'VenueName','VenueLatitude', 'VenueLongitude', 'VenueCategory']
print(venues_df.shape)
venues_df.head()
print(venues_df)
venues_df.groupby(["CityName"]).count()
print(venues_df.groupby(["CityName"]).count())
```

```python
print('There are {} uniques categories.'.format(len(venues_df['VenueCategory'].unique())))
# print out the list of categories
venues_df['VenueCategory'].unique()[:50]
```

```python

# one hot encoding
kl_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add CityName column back to dataframe
kl_onehot['CityName'] = venues_df['CityName'] 

# move CityName column to the first column
fixed_columns = [kl_onehot.columns[-1]] + list(kl_onehot.columns[:-1])
kl_onehot = kl_onehot[fixed_columns]

print(kl_onehot.shape)
kl_onehot.head()
print(kl_onehot)
```

```python

kl_grouped = kl_onehot.groupby(["CityName"]).mean().reset_index()

print(kl_grouped.shape)
kl_grouped
print(kl_grouped)
```

```python

#Create a new DataFrame for Housing Development data only
len(kl_grouped[kl_grouped["Housing Development"] > 0])

kl_mall = kl_grouped[["CityName","Housing Development"]]
kl_mall.head()
print(kl_mall)
```

```python

# set number of clusters
kclusters = 3
kl_clustering = kl_mall.drop(["CityName"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(kl_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]

# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
kl_merged = kl_mall.copy()
```

```python
# add clustering labels
kl_merged["Cluster Labels"] = kmeans.labels_

kl_merged.rename(columns={"CityName": "CityName"}, inplace=True)
kl_merged.head()
print(kl_merged)

kl_merged = kl_merged.join(dfsrt.set_index("CityName"),on="CityName")
print(kl_merged.shape)
kl_merged.head() # check the last columns!
print(kl_merged)

print(kl_merged.shape)
kl_merged.sort_values(["Cluster Labels"], inplace=True)
kl_merged
print(kl_merged)
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
```

```python

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(kl_merged['LATITUDE'], kl_merged['LONGITUDE'], kl_merged['CityName'], kl_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)       
map_clusters
map_clusters.save('map_clusters.html')
```
